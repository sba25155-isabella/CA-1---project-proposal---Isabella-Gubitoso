#Data Preparation and EDA visualization

In this section I will clean the dataset by fixing data quality issues and then create visualizations to better understand the patterns in the student data provided.

Before cleaning, there were seven differnt inconsistent data entry. I mapped all variations to three correct categories: Graduate, Dropout and Enrolled.

After cleaning, there are only three categories with no missing values.

```python
#Dropping the missing values
df=df.dropna()
df.count()
```

This code remove rows that contain missing values. Before cleaning, there were nine columns with one missing value each. After dropping these rows, it's removed only a few students from the dataset, which will not affect the analysis results beacuse we are losing less than 0,1% of the data.

this cleaning is necessary because machine learning models require consistent category labels for proper analysis.

#PCA

Now, I apply PCA to reduce the dimensionality of the dataset. I want to find the minimum number of components that exlain 99.5% of variance.

```python
from sklearn.decomposition import PCA
import numpy as np
```

```python
numeric_features = df.select_dtypes(include=['int64', 'float64'])
if 'Target' in numeric_features.columns:
    numeric_features = numeric_features.drop('Target', axis=1)
numeric_features = numeric_features.dropna()
X = numeric_features.values
pca = PCA(n_components=2)
pca = PCA()
pca.fit(X)
PCA(copy=True, n_components=2, whiten=False)
```

```python
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)
n_components_995 = np.argmax(cumulative_variance >= 0.995) + 1
print(f"Number of features: {X.shape[1]}")
print(f"Components for 99.5% variance: {n_components_995}")
print(f"Dimentionality reduction: {X.shape[1] - n_components_995} features removed")
print(f"Actual variance explained: {cumulative_variance[n_components_995-1]*100}%")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components', fontsize=12)
plt.ylabel('Cumulative Explained Variance', fontsize=12)
plt.title('Cumulative Variance Explained', fontsize=14, fontweight='bold')
plt.grid(alpha=0.3)
plt.show()
```

This plot shows how cumulative variance increases as the curve rises in the first 5-10 components. After that, the curve flatterns.

#The curse of dimensionality

The curse of dimensionality is when you have too many features in your data. When there are many dimensions, data points become very spread out in space. This makes it difficult to find patterns and relationship. Machine learning algorithms need much more data to learn properly, and they become slower.

This dataset has 31 features, which means is high dimensional data. Many of these features are correlated, giving similar information. 

PCA solves this problem by reducing dimensions. In this dataset, it found that only 5 to 10 can explain 99,5% of the variance.This means I can use much fewer features and still keep almost all the information.

By reducing from 31 to 5-10 dimensions, the data becomes easier to work with machine learning algorithms, and it performs better and faster. And that is the reason of the importance of PCA for high dimensional datasets.
